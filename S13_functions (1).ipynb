# Functions

In programming, a function is a block of instructions whose objective is to perform a specific task. It differs from a normal sequence of instructions by being **reusable**.

## Rationale
Functions are very popular because 
- they allow us to **reuse** the same instructions many times, in many different places, without having to repeat them every time we need them
- they allow us to divide very long or complex tasks into smaller, more specific tasks, and by encapsulating each task in a different function, improve the readability of our programs.
- having the definition in a single **centralized** place it makes code much easier to debug.


# Definition vs Call
Functions have two *parts*:
1. the **definition**  

2. the **call**  

To be able to *call* a function, we must have defined it (or imported it) **beforehand**. A function can be called an infinite amount of time in many different places, but it must be defined only once!

> **FYI** - **PRE-DEFINED FUNCTIONS**  
Up until now we only used pre-defined functions (i.e. functions defined somewhere). Indeed, the Python language already provides some functions so that we do not have to write them ourselves from scratch. In that case we either:
>- used them directly, because they're **built-in**; some examples might be `len`, `sum` or `sorted`
>- imported them from a library; for example, we used the function `randomint` from the package `random`
>
> From now on we will be able to define our own functions, too.


## Definition

The first step to be able to use a function is to **define** it.   
This is the step in which we
- define a name for the function
- define which inputs it expects and which outputs it returns
- define the core of the logic

To some extent, we can see the definition as an assignment-like operation, with the difference that in this case we are not going to store in memory a simple value, rather a sequence of instructions.


### Structure

The structure of a function is as follows:

- **`def`**  
   it is a **mandatory** keyword that tells that we are going to define a function next.

- **name**  
   an identifier (chosen by us - the same rules for variable identifiers apply) for the function.

- **parameters**  
   a sequence of identifiers (defined by us) within parentheses and separated by commas, indicating the values that the function expects as input.  
   > **IMPORTANT:** Parameters are optional, but parentheses are not. If the function doesn't need parameters, the parentheses will simply remain empty.

- **body**  
   a block of instructions to be executed when the function is invoked. In this body we can make use of the variable(s) specified as input of the function.
   > **REMEMBER:** The block must always be indented one level from its definition.

- **`return <expression>`**  
   a reserved keyword indicating that the execution of the function should end. 
   - There can be 0 or more lines of `return`. 
      - If there isn't a `return` statement, the function terminates as soon as the execution reaches its last line, and the output will be automatically `None`.
      - If there is one or multiple `return` statements, the execution of the function terminates as soon one of these statements is reached.
   - Optionally, you can attach an expression to this keyword, such as `return <expression>`. In that case, the function will return `<expression>` as an output. If no value is specified, the `return` statement automatically returns `None`.


<center></center>

````python
def bmi(height_in_cm, weight_in_kg):
    """
    This function returns the BMI given 
    the height in cm and the weight in kg
    """

    height_in_m = height_in_cm / 100
    bmi = weight_in_kg / height_in_m ** 2

    return round(bmi, 1)
````

# Parameters vs Arguments

Parameters and arguments are technical words that both refer to function **inputs**.

- With **parameters**, we indicate the inputs in the **definition** of a function. Parameters are the **variables** that the function expects in input.
- With **arguments**, we refer to the inputs during the **call**. Arguments are part of the call. Arguments are the **values** that we *pass* to to the functions.  



By default, a function must be invoked with as many arguments as there are parameters defined. For example, if we define `bmi` with two parameters, we must always invoke `bmi` with exactly two arguments, neither more nor less.

````python
# The function bmi was defined with two parameters, therefore we cannot
# invoke it with fewer than two arguments...
bmi(180)
````

````python
# The function bmi was defined with two parameters, therefore we cannot
# invoke it with fewer than two arguments...
bmi()
````

````python
# ...nor with more!
bmi(180, 80, 100)
````

## Positional and Keyword Arguments

````python
def foo(x, y, z):
    """Just a dumb function..."""
    print(f'x: {x}; y: {y}; z: {z};')
````

### Positional Arguments

**By default, arguments are *positional***. What does it mean?  
It means that by default, when I pass some *arguments* to a defined function, the mapping between *arguments* and *parameters* is done **implicitly** based on the **position** of each argument during the call. That is, the first argument will be assigned to the first defined parameter, the second argument to the second parameter, and so on...  

For example, if a function was *defined* with 3 *parameters* `x`, `y`, `z`, and then the function is *called* with the arguments `4`, `10`, `9`, the implicit association between parameters and arguments will be:
- `x` = `4`
- `y` = `10`
- `z` = `9`

In other words, positional arguments are those arguments that are passed *by position* (i.e., whose corresponding parameter can be inferred from their position).


````python
# If we pass arguments by position, the order counts!
foo(4, 10, 9)

foo(10, 4, 9)

foo(9, 10, 4)
````

### Keyword Arguments

On the other hand, there is another more explicit option for passing arguments: passing them **by keyword**.  

Passing an argument *by keyword* means specifying the argument along with its identifier/parameter. To do that, you can use the following syntax:  
`<parameter>=<argument>`  

This allows us to explicitly specify the parameter-argument relationship without having to deduce it from the position.
Similar to the case of positional arguments, arguments passed by keyword are called *keyword arguments*.

> **IMPORTANT**  
If we pass an argument by keyword, there is no doubt about which parameter that argument refers to. Therefore, when passing arguments by keyword, it is **not obligatory** to respect the original order of the parameters.


````python
# If we pass arguments by keyword,
# we can maintain the same order of the arguments...
foo(x=4, y=10, z=9)

# ...or change it!
foo(y=10, x=4, z=9)
````

````python
# we cannot pass a keyworded argument twice!
foo(y=10, y=9, x=4)
````

### Precedence of Positional Arguments over Keyword Arguments

Another possibility is to "mix" the two ways, that is, to pass some arguments *by position* and others *by keyword*.  
In this case, **positional arguments must always precede keyword arguments**. If we try to first pass some parameters by keyword followed by other positional parameters, the interpreter will throw a `SyntaxError`.


````python
# here we are passing some arguments by position, and other by keyword
# in this case x=4 is inferred
foo(4, z=10, y=9)
````

````python
# here we are passing some arguments by position, and other by keyword
# in this case, x=4 and y=9 are inferred
foo(4, 9, z=10)
````

````python
# here we are passing some arguments by position, and other by keyword
# BUT we are placing the keyworded args first, which raises an error!
foo(x=4, 9, 10)
````

## Default Arguments

Sometimes, in the definition of a function, we want to specify a default value for one or more parameters. We can do it by indicating the default value in the definition like so:

````python
# in this case z has a default value, which is 10
def baz(x, y, z=10):
    print(f'{x=} {y=} {z=}')

# here I can pass 2 positional arguments although the function has 3 parameters
# because one parameter has a default value!
baz(1, 2)
````

## Error Handling: `try` / `except`

When we write a piece of code we must consider that probably someone else will later run it. Sometimes we can't control for all the possible scenarios and all the possible inputs, yet we want to make sure that our program handles any kind of situation without crashing.  

Python provides us with a construct that allows us to do that: `try` / `except`.

In the `try` we will place the piece of code that might raise some unexpected exception at runtime, in the `except` we will say what the program needs to do whenever an Exception is raised.

The syntax is as follows:

```python
try:
    # anything you want to try to execute
    ...
except Exception:
    # anything you want to execute in case of an error
    ...

````python
# if the user does not input a numerical value, the code crashes
age = int(input('Your age:'))
````

````python
while True:
    try:
        # if the user does not input a numerical value, the code crashes
        age = int(input('Your age:'))
        break
    except Exception:
        print(f'The input value is not numerical!')
        continue
````

````python
"""
If possible, you should always prefer implementing a robust error handling strategy
instead of using try/except as a lifesaver.
"""

while True:
    age = input('Your age:')
    # in this case I don't really need a try/except because 
    # I can already control for all the "exceptional" cases
    if not age.isnumeric():
        continue
    age = int(age)
    break
````



## Scope

In Python, like in many other programming languages, each object has its own *scope*. We can think of scope as a space, an environment within which the object "exists" and which defines which other objects it can see and interact with.


### Intuition
- In general, everything defined at the top level of a Python program is part of the same environment called ***global***. The global environment is **unique**.
- However, everything defined within functions is not part of the global environment but rather of separate environments for each function called **local** environments. This is why functions force us to talk about *scopes*.


### Rules
- **Global** objects are accessible from any environment, global or local. In other words, what we define at the top level of a Python program is accessible from anywhere, even from within a function or a nested function.
- **Local** objects are only accessible within their own environment and from nested local environments, but not from the global environment or from other unrelated local environments. In other words, what we define within a function is not available outside the function or in other functions, but only within the same function and within any possible nested functions.
- Although objects from more "general" environments can be used in more "local" environments, they can only be modified within their own environment. In other words, we cannot modify a global variable within a function.

````python
my_name = 'Massimiliano'

def bar():
    # in this case my_name exists in the global space, so I can use it
    print('My name is', my_name)

bar()
````

````python
"""
Let's see that we cannot access local variables from the global environment,
but we can access global variables from any local environment.
"""

# Define some global variables
global_greeting = 'Hello!'

def send_local_greetings():
    # Everything defined here is part of a local environment and exists only
    # within this scope and within any possible nested functions
    
    # Define other variables 
    # this variable is now local!
    local_greeting = 'Merhaba!'
    
    # We can access both the local and global greetings 
    # because global variables are accessible from any scope
    print("From a local environment:")
    print(f"- global greeting: {global_greeting}")
    print(f"- local greeting: {local_greeting}")
    
    
# Execute the function that sends us local greetings
send_local_greetings()

print()

# However, we cannot do the same greetings from the global environment
# because the global environment cannot see variables from local scopes
print("From the global environment:")
print(f"- global greeting: {global_greeting}") 
print(f"- local greeting: {local_greeting}") # This will raise an error
````

````python
"""
Let's prove that a local environment cannot modify objects from a global environment.
Although it may seem that the object is being modified, what is actually being modified
is a local "copy".
"""

# let's define a global variabel
global_greeting = 'Hello!'

def replace_global_greetings():
    
    local_greeting = 'Merhaba!'
    # let's try to replace the global value with a local value
    # (this line actually does not modify the global value, 
    # simply creates a local variable that has the same name)
    global_greeting = local_greeting
    
    # Let's see if it works...
    print("From a local environment:")
    print(f"- local greeting: {local_greeting}")
    print(f"- global greeting: {global_greeting}")
    
    
# Execute the function that sends us local greetings
# (it seems like the plan works...)
replace_global_greetings()

print()

print("From the global environment:")
# Almost! Luckily, the global variable is actually left untouched!
print(f"- global greeting: {global_greeting}") 
````

````python
"""
To succesfully modify the value of a global variable from a local environment,
then we have to specify that we are referring to a global variable by stating
global <variable_name>
somewhere within the local environment.
"""

# let's define a global variabel
global_greeting = 'Hello!'

def actually_replace_global_greetings():
    
    local_greeting = 'Merhaba!'
    # let's try to replace the global value with a local value
    # if I use the global keyword, then I am actually modifying the global version!
    global global_greeting
    global_greeting = local_greeting
    
    # Let's see if it works...
    print("From a local environment:")
    print(f"- local greeting: {local_greeting}")
    print(f"- global greeting: {global_greeting}")
    
    
# Execute the function that sends us local greetings
# (it seems like the plan works...)
actually_replace_global_greetings()

print()

print("From the global environment:")
# Almost! Luckily, the global variable is actually left untouched!
print(f"- global greeting: {global_greeting}") 
````
