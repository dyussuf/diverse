```python
""" EXERCISE 13.1

Write a program in a .py file which calculate the volume for 3 possible shapes: a cube, a cylinder and a cone. 

The formulas are:
- Cube =        $s^3$		            
    (s = one side)
- Cylinder =    $π r^2 h$		        
    (r = radius, h = height)
- Cone =        $π r^2 \frac{h}{3}$  
    (r = radius, h = height)

Rules:
- There must be a `main()` function which starts with a menu asking which calculation to do, e.g.:
> Choose from the following options:  
>- Cube  
>- Cylinder  
>- Cone   

- Create three functions (one for each calculation) which will be later called by the `main()` function.

- The application should then get the appropriate inputs and display the result of the chosen calculation.

- Handle the situation in which the user indicates a shape that is not supported by your program.
"""
import math

def get_cube_area(s):
    return s ** 3

def get_cylinder_area(r, h):
    return math.pi * (r ** 2) * h

def get_cone_area(r, h):
    return math.pi * (r ** 2) * (h / 3)

def main():

    print('This program allows you to compute the area of a given shape. The valid shapes are:')
    print('- CUBE')
    print('- CYLINDER')
    print('- CONE')

    while True:
        selected_shape = input('Choose the shape for which you need to compute the area (type the shape uppercase).')
        print(f'You selected: {selected_shape}')

        if selected_shape == 'CUBE':
            s = float(input('Specify the side of the cube: '))
            area = get_cube_area(s=s)
            print(f'The area of a cube with side={s} is {area}.')
            return area
            
        elif selected_shape == 'CYLINDER':
            r_h = input('Specify the radius and the height of the cylinder, separated by comma: ')
            r, h = r_h.split(',')
            r = float(r)
            h = float(h)
            area = get_cylinder_area(r=r, h=h)
            print(f'The area of a cylynder with radius={r} and height={h} is {area}.')
            return area
        
        elif selected_shape == 'CONE':
            r_h = input('Specify the radius and the height of the cone, separated by comma: ')
            r, h = r_h.split(',')
            r = float(r)
            h = float(h)
            area = get_cone_area(r=r, h=h)
            print(f'The area of a cone with radius={r} and height={h} is {area}.')
            return area

        else:
            print(f'Your selection {selected_shape} is invalid. Retry...')

main()
```

```python
""" EXERCISE 13.2

Create a function `random_birthdate` that generates and returns a random birthdate in the form of a tuple `(day, month)`. 
For example `(22, 3)` to indicate 22nd of March.

Now generate random birthdates for N people. 
This functionality should be encapsulated inside another function called `random_n_birthdates` 
that given a parameter `N` returns a list with N random birthdates.

Now generate a program that makes use of these functions. 
The idea of the program is that given a number N in input, it returns whether there are at least two people 
that share the same birthdate in a group on N people. 
The output of this function should be either an integer 0/1 or a boolean True/False.

"""
import random

MAX_DAYS = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def random_birthdate() -> tuple:
    # let's pick a random month between 1 and 12
    month = random.randint(1, 12)
    # once the month is fixed, let's pick a random day between 1 and the maximum number of days for that month
    day = random.randint(1, MAX_DAYS[month-1])
    # once we have the day and the month, let's build and return the date tuple
    return (month, day)


def random_n_birthdates(n) -> list:
    # let's initialize an empty list
    birthdates = []
    # let's add a new random date per n times
    for _ in range(n):
        birthdates.append(random_birthdate())
    # let's return the list with all the dates
    return birthdates


def main(n=60):
    """Here we want to say whether there are birthdays in common..."""

    birthdates = random_n_birthdates(n=n)

    # let's iterate over each birthday 
    # (there is no need to iterate over the last one because there is nobody to compare against)
    for ith_birthdate, birthdate_i in enumerate(birthdates[:-1]):

        # now we must compare each birthday to the "next" ones
        for jth_birthdate, birthdate_j in enumerate(birthdates[ith_birthdate+1:]):
            
            # if the birthdays are different, we keep on comparing...
            if birthdate_i != birthdate_j:
                continue
            
            # otherwise the birthdays are equal and so we can already terminate
            # saying True, there is at least one birthday in common
            return True
    
    # by default we can say there are no birthdays in common
    return False

main()
```
