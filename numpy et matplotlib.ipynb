# Le module numpy

## Introduction

$numpy$ $(numerical\>python)$ est un module Python optimisé pour effectuer des calculs mathématiques de manière rapides et plus performante sur les vecteurs et les matrices.

Commençons par importer le module et lui donner l’alias $np$ afin de faciliter son utilisation :

```python
import numpy as np   
```

## Les fonctions usuelles

Dans le module $numpy$ on en trouve plusieurs comme :

-  Fonctions trigonométriques : $np.cos$, $np.sin$, ...

-  Fonctions logarithmiques et exponentielles : 

   - $np.log$ pour le logarithme naturel;
   
   - $np.log2$ et $np.log10$ pour le logarithme à base 2 et à base 10, ...
   
   - $np.exp$ pour l'exponentielle, ...
   
-  La fonction $np.sqrt$ qui renvoie la racine carrée et $np.absolute$ pour la valeur absolue, ...

## Les vecteurs

Pour définir un vecteur ligne, on utilise le mot clé $array$ suivi par les coordonnées du vecteur.

### Exemple

```python
vecteur = np.array([1, 3, 7])   #crée un vecteur de coordonnées (1, 3, 7)
```

## Créaction et utilisation rapide de vecteurs

- $np.arange(début, fin, pas)$ renvoie un vecteur ligne dont les valeurs sont en progression arithmétique de raison $pas$, de premier terme $début$ et ne dépassant pas $fin$.

- Si on a besoin de générer un vecteur de nombres uniformément espacés sur un intervalle [début, fin] on peux utiliser np.linspace(début, fin, nombre_élément). 

- $np.ones$ renvoie un vecteur ligne dont tous les coefficients sont égaux à 1.

- $np.zeros$ renvoie un vecteur ligne dont tous les coefficients sont égaux à 0.

### Exemples

```python
v1 = np.arange(1, 10, 0.5)   #un vecteur de valeurs d'une suite arithmétique de raison 0.5 à partir de 1 et < 10

print(v1)
```

```python
v2 = np.linspace(1, 10, 20)   #un vecteur de 20 valeurs sur [1, 10]

print(v2)
```

Généralement, $arange$ est utilisée lorsqu'on connaît le pas et $linspace$ lorsqu'on connaît le nombre de valeurs.

```python
v3 = np.ones(7)   #un vecteur contenant sept un

print(v3)
```

```python
v4 = np.zeros(10)   #un vecteur nul de dimension 10

print(v4)
```

```python
# Accès à une valeur

v = np.array([1,2,8])
print(v[0])  #élément d'indice 0
```

```python
# Modification d'une valeur 

v[1] = 3
print(v)
```

## Image d'un vecteur par une fonction

Pour obtenir un vecteur d'images de valeurs d'un vecteur par une fonction il suffit de donner ce vecteur comme argument à cette fonction.

### Exemple

```python
def inverse(x):
    return 1/x

v = np.array([1,2,8])

print(inverse(v))
```

# Le sous module matplotlib.pyplot

## Introduction

$matplotlib.pyplot$ est une bibliothèque $Python$ pour créer des visualisations de données sous forme de graphiques comme les courbes, les nuages de points, les diagrammes en bâtons, les histogrammes, ...

Commençons par l'importer sous l'alias $mpl$ :

```python
import matplotlib.pyplot as mpl
```

## Courbe d'une fonction

Après avoir créé les données (vecteur d'abscisses et vecteur d'ordonnés) , on fait appel à la fonction $plot(abscisses, ordonnées)$ qui permet de tracer une ligne brisée continue entre les points donnés.

### Exemple

Traçons la courbe de $f(x)=\frac{1}{x}$ sur [1, 10] :

```python
# définition de la fonction
def f(x):
    return 1/x

# création des abscisses 
x = np.linspace(1, 10, 50)   #spécifie que nous voulons 50 valeurs sur [1, 10].

# création des ordonnées
y = f(x)

# Traçage de la courbe
mpl.plot(x, y)

#Ajouter une grille
mpl.grid()

# ajouter une légende aux axes
mpl.xlabel('x')    
mpl.ylabel('1/x')

# ajouter un titre
mpl.title('inverse')
```

$REMARQUE$ : Plus le nombre des abscisses est grand, plus la courbe est précise et approchera mieux la forme mathématique de la fonction. 

Pour tracer plusieurs courbes sur la même figure il suffit d'ajouter autant de plot que nécessaire et de personnaliser de préférence chaque courbe par :

- un label en utilisant le paramètre $label$ pour l'identifier dans la légende;

- une couleur avec le paramètre $color$ et des noms de couleurs en anglais;

- un style de ligne avec le paramètre $linestyle$ :

   - $-$ pour une ligne continue (valeur par défaut)
   - $--$ pour une ligne discontinue 
   - $-.$ pour une alternance de points et tirets
   - $:$ pour une ligne pointillée 
   
### Exemple

Traçons les courbes des fonctions $sinus$ et $cosinus$ sur $[-2 \pi, 2 \pi]$ :

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)  
y1 = np.sin(x)
y2 = np.cos(x)
mpl.plot(x,y1, label='sin', color='red', linestyle='-.')
mpl.plot(x,y2, label='cos', color='green', linestyle='--')

mpl.legend()   #pour ajouter une légende avec les noms des courbes
```

Pour afficher chaque courbe séparément, il suffit de créer une nouvelle figure avec la fonction $figure()$ avant le tracement de chaque courbe :

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100) 

#Figure 1 : sinus
mpl.figure()
mpl.plot(x, np.sin(x), label='sin', color='red')
mpl.title('sinus')

# Figure 2 : cosinus
mpl.figure()
mpl.plot(x, np.cos(x), label='cos(x)', color='green')
mpl.title('cosinus')

# Afficher toutes les figures
mpl.show()
```

On peut même ajouter des annotations sur la figure avec la fonction $annotate$ en précisant le text à écrire et ses coordonnées :

### Exemple

```python
def opposé(x):
    return -x

x = np.arange(-10, 11, 1)
mpl.plot(x, opposé(x))

mpl.annotate('Élément neutre', (0, 0))   #écrire 'Élément neutre' aux coordonnées (0, 0)

mpl.xlabel('x')    
mpl.ylabel('-x')
mpl.grid()
mpl.title('opposé')
```
