## Tri par slection

Le tri par slection est un tri en place et non stable. Ce tri fonctionne en slectionnant l'lment le plus petit de la partie non trie et en le permutant avec le premier lment de cette partie. Il continue ensuite avec le reste des lments (en ignorant la partie dj trie) jusqu' ce que tout soit tri.

```python
def tri_par_slection(liste):
    for i in range(len(liste)-1):
        indice_minimum = liste.index(min(liste[i:]))
        liste[i], liste[indice_minimum] = liste[indice_minimum], liste[i]
        print(liste)
    return liste

liste = [5,1,7,2,6,4,3]
tri_par_slection(liste)
```

## Tri par insertion

Le tri par insertion est effectu  en insrant chaque lment reu  sa bonne position dans la partie trie. Il fonctionne comme lorsque l'on tient des cartes  jouer : on prend sur la table les cartes, une par une, pour l'insrer  la bonne position dans la main gauche en examinant les cartes, dj tries, de la droite vers la gauche.

### Exemple de code

```python
def tri_par_insertion(liste):
    for i in range(1, len(liste)):
        to_insert = liste[i]
        where = i-1
        while to_insert < liste[where] and where >= 0:
            liste[where+1] = liste[where]
            where -= 1
        liste[where+1] = to_insert
        print(liste)
    return liste

liste = [5,1,7,2,6,4,3]
tri_par_insertion(liste)
```

Le tri  bulles sappuie sur des permutations rptes dlments contigus qui ne sont pas dans l'ordre afin de placer le plus grand lment  sa position finale. Ce processus est rpt (en ignorant la partie dj trie) jusqu' ce que tout soit tri.

```python
def tri__bulles(liste):
    for i1 in range(len(liste)-1):
        changed = False   # flag pour optimiser, la liste est trie si aucune permutation n'est dtecte.
        for i2 in range(len(liste)-1-i1):
            if liste[i2]>liste[i2+1]:
                liste[i2], liste[i2+1] = liste[i2+1], liste[i2]
                changed = True
        print(liste)
        if not changed:
            break
                
liste = [5,1,7,2,6,4,3]
tri__bulles(liste)
```

# Tri par comptage

Le tri par comptage est un algorithme de tri non comparatif utilis principalement lorsque les valeurs  trier sont dans un domaine de nombres limit. Il est trs efficace lorsque l'cart entre la plus petite et la plus grande valeur est raisonnable par rapport  la taille des donnes.

$Principe$ :

- Trouver la valeur minimale et la valeur maximale.

- Crer une liste de comptage o chaque indice reprsente une valeur possible de la liste originelle, et stocker le nombre doccurrences de chaque valeur.

- Parcourir la liste de comptage pour reconstruire la liste trie.

## Exemple de code 

```python
def tri_par_comptage(liste):
    mini, maxi = min(liste), max(liste)
    comptage = [0 for _ in range(maxi-mini+1)]
    for i in range(mini, maxi+1):
        comptage[i-mini] = liste.count(i)
    print('comptage =', comptage)
    rsultat = []
    for x in comptage:
        rsultat.extend(x*[mini])
        mini += 1
    return rsultat

liste = [1,1,1,3,3,5,7,7,7,7]
tri_par_comptage(liste)
```
